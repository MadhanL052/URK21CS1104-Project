```python
from ultralytics import YOLO

# Load a YOLO model
model = YOLO("yolov5n6u.pt")  # Pre-trained YOLOv8 model

# Train the model on your dataset
model.train(
    data="/kaggle/input/ydataset1000/yolo_dataset_1000/data.yaml",  # Dataset configuration
    epochs=250,                     # Number of training epochs
    imgsz=1280,                     # Image size
    batch=16,                      # Batch size
    device=0                       # GPU device index (0 for the first GPU)
)
```

```python
metrics = model.val(data="/kaggle/input/ydataset1000/yolo_dataset_1000/data.yaml", batch=16, imgsz=1280)

# Access metrics from the 'metrics.box' attribute
precision = metrics.box.p.mean()  # Mean Precision
recall = metrics.box.r.mean()     # Mean Recall
map50 = metrics.box.map50         # Mean Average Precision at IoU=0.5
map50_95 = metrics.box.map        # Mean Average Precision at IoU=0.5:0.95

# Display metrics
print("Validation Results:")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"mAP@0.5: {map50:.4f}")
print(f"mAP@0.5:0.95: {map50_95:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load results from YOLO training logs
csv_path = "/kaggle/working/runs/detect/train/results.csv"  # Adjust this based on your training folder
df = pd.read_csv(csv_path)

# Plot training and validation loss
plt.figure(figsize=(10, 5))
plt.plot(df["epoch"], df["train/box_loss"], label="Train Box Loss")
plt.plot(df["epoch"], df["train/cls_loss"], label="Train Class Loss")
plt.plot(df["epoch"], df["val/box_loss"], label="Validation Box Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.show()

# Plot mAP metrics
plt.figure(figsize=(10, 5))
plt.plot(df["epoch"], df["metrics/mAP50(B)"], label="mAP@0.5")
plt.plot(df["epoch"], df["metrics/mAP50-95(B)"], label="mAP@0.5:0.95")
plt.xlabel("Epochs")
plt.ylabel("mAP")
plt.title("mAP Over Epochs")
plt.legend()
plt.show()
```
